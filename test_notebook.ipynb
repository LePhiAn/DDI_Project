```python
import pandas as pd

file_path = 'data/raw/ChChSe-Decagon_polypharmacy.csv'

# Đọc lại file với các tham số điều chỉnh:
# 1. skiprows=1: Bỏ qua dòng đầu tiên có chứa dấu #
# 2. names=[...]: Đặt lại tên cột cho sạch sẽ
# 3. sep=',': Đảm bảo dùng dấu phẩy để tách cột
df = pd.read_csv(
    file_path, 
    sep=',', 
    skiprows=1, 
    names=['Drug1', 'Drug2', 'SideEffect_ID', 'SideEffect_Name'],
    low_memory=False
)

# Xóa bỏ các khoảng trắng thừa (nếu có) ở đầu/cuối mỗi ô dữ liệu
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

print("✅ Đã tách cột thành công!")
print(df.head())

print(f"\nTổng số cặp tương tác: {len(df):,}")
print(f"Số lượng loại tác dụng phụ: {df['SideEffect_Name'].nunique()}")
```
